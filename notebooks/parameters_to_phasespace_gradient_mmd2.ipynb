```python
@jax.jit
def rbf_kernel(x, y, sigma):
    """RBF kernel optimized for 6D astronomical data"""
    return jnp.exp(-jnp.sum((x - y)**2) / (2 * sigma**2))
```

```python
import jax.numpy as jnp
from jax import random, value_and_grad, jit
import optax

# --- ASSUMPTION: BATCH_SIZE and squared_distance_matrix ---
# Define a memory-safe batch size
BATCH_SIZE = 512 

# Ensure this function (from previous discussions) is defined and available
@jit
def squared_distance_matrix(A, B):
    """Computes ||A_i - B_j||^2 matrix for all pairs of vectors."""
    A_sq_norm = jnp.sum(A**2, axis=1, keepdims=True)
    B_sq_norm = jnp.sum(B**2, axis=1, keepdims=True).T
    dot_product = jnp.dot(A, B.T)
    D_sq = A_sq_norm + B_sq_norm - 2 * dot_product
    return jnp.maximum(D_sq, 0.0)

# The full simulation function MUST be checkpointed for memory robustness
@jax.jit
@jax.checkpoint # Crucial for complex simulation memory management
def params_to_phasespace_compiled(params, theta=theta, Phi= Phi, key=key, n_candidates=n_candidates, envelope_max=envelope_max):
    # ... (Your existing params_to_phasespace logic here) ...
    # Placeholder to ensure signature is known:
    return params_to_phasespace(params, theta, Phi, key, n_candidates, envelope_max)


# --- SUBSAMPLED MMD LOSS FUNCTION ---
# NOTE: The signature must now include 'key' for stochastic subsampling.
@jax.jit
def loss_function(params, theta_init, theta_target): 
    # 1. Generate FULL phase spaces (N particles)
    # Split key for simulation (sampling angles) and MMD subsampling
    key_sim, key_mmd = random.split(key)

    theta = jnp.array([theta_init[0]*1e12, theta_init[1]*20, theta_init[2]*20])
    theta_target = jnp.array([theta_target[0]*1e12, theta_target[1]*20, theta_target[2]*20])
    
    # Use key_sim for stochastic sampling inside the forward model
    guess_phase_space = params_to_phasespace_compiled(params, theta=theta_init, Phi= Phi, key=key, n_candidates=n_candidates, envelope_max=envelope_max)
    target_phase_space = params_to_phasespace_compiled(params, theta=theta_target, Phi= Phi, key=key_sim, n_candidates=n_candidates, envelope_max=envelope_max)

    # 2. Subsample phase spaces (B particles)
    N_guess = len(guess_phase_space)
    
    key_idx_1, key_idx_2 = random.split(key_mmd)

    # Draw BATCH_SIZE indices from N_guess/N_target
    idx_guess = random.choice(key_idx_1, N_guess, shape=(BATCH_SIZE,), replace=False)
    idx_target = random.choice(key_idx_2, N_guess, shape=(BATCH_SIZE,), replace=False)

    guess_batch = guess_phase_space[idx_guess]
    target_batch = target_phase_space[idx_target]

    # 3. Recalculate sigma based on BATCH_SIZE
    sigma = 0.5 * jnp.power(2 * BATCH_SIZE, -1/(6+4))

    # 4. Calculate MMD on the B x B matrices
    D2_guess = squared_distance_matrix(guess_batch, guess_batch)
    D2_target = squared_distance_matrix(target_batch, target_batch)
    D2_cross = squared_distance_matrix(guess_batch, target_batch)

    K_guess = jnp.exp(-D2_guess / (2 * sigma**2))
    K_target = jnp.exp(-D2_target / (2 * sigma**2))
    K_cross = jnp.exp(-D2_cross / (2 * sigma**2))

    loss = jnp.mean(K_guess) + jnp.mean(K_target) - 2 * jnp.mean(K_cross)
    
    return loss
```

```python
# Updated step logic to handle the new loss_function signature
def step_logic(params, theta, theta_target, optimizer_state, optimizer, key): 
    # Use a lambda to correctly pass both theta (argnums=0) AND key (non-differentiated)
    loss, grads = value_and_grad(
        lambda t, k: loss_function(params, t, theta_target, k), 
        argnums=0
    )(theta, key)
    
    grads = jnp.clip(grads, -1e3, 1e3)
    updates, optimizer_state = optimizer.update(grads, optimizer_state, theta)
    theta = optax.apply_updates(theta, updates)
    return theta, loss, grads, updates, optimizer_state

# The run_adam_optimizer must now accept the initial key
def run_adam_optimizer(loss_function, params, theta_init, theta_target, initial_key, learning_rate=1e-3, tol=1e-6, max_iter=10):
    # Setup
    optimizer = optax.adam(learning_rate)
    optimizer_state = optimizer.init(theta_init)
    theta = theta_init
    key = initial_key # Start using the key
    
    theta_history = []
    loss_history = []

    for i in range(max_iter):
        # CRITICAL: Split key for each stochastic step
        key, subkey = random.split(key)

        theta, loss, grads, updates, optimizer_state = step(
            params, theta, theta_target, optimizer_state, optimizer, subkey
        )

        loss_history.append(loss)
        theta_history.append(theta)
        
        print(f"Iteration {i}: Loss = {loss}, Theta = {theta}")

        if optax.global_norm(updates) < tol:
            print(f"Converged at iteration {i}")
            break
    
    return theta, loss, theta_history, loss_history

# Example Initialization:
# initial_key = jax.random.PRNGKey(42)
# theta_optimized, final_loss, theta_history, loss_history = run_adam_optimizer(
#     loss_function, params, theta_init, theta_target, initial_key, learning_rate=1e-6, max_iter=500
# )
```
